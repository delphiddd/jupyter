```python
import pandas as pd
```

```python
df = pd.read_csv('data/simple_loan.csv')
```

```python
df
```

LABEL ENCODING

```python
def LabelEncode(data , col):
    for i in col:
        data[i +'_'] , _ = pd.factorize(data[i] , sort=True)
feature_cols = ['age' , 'employed' , 'own_house' , 'credit' , 'target']
LabelEncode(df , feature_cols)
```

```python
df
```

```python
X = df[['age_', 'employed_', 'own_house_', 'credit_']]
```

```python
y = df[['target_']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=20) 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train , y_train)
```

```python
score = model.score(X_test , y_test)
print(score)
```

```python
y_pred = model.predict(X_test)
X_test
```

```python
print(y_pred)
```
